```python
import json
import matplotlib.pyplot as plt
from collections import defaultdict


# Khởi tạo các list để lưu số lượng từ trong câu hỏi, câu trả lời và đoạn văn
num_words_questions = []
num_words_answers = []
num_words_paragraphs = []
num_words_distractor = []
exist_ques_id = defaultdict(int)
number_ques_per_paragraph_dict = defaultdict(int)
number_ques_per_paragraph = []


def convert_json_file(file_path):
    # Step 1: Read the file content
    with open(file_path, 'r', encoding='utf-8') as dict_file:
        # with open(file_path, 'r', encoding='utf-8') as file:
            for line in dict_file:
                content = json.loads(line)
                answer_text = content.get('answer_text')
                num_words_answers.append(len(answer_text))
                distractor = content.get('distractor')
                num_words_distractor.append(len(answer_text))
                question = content.get('question')
                num_words_questions.append(len(question))
                file_id = str(content.get('id')['file_id'])
                ques_id = str(content.get('id')['file_id']) + '_' + str(content.get('id')['question_id'])
                if exist_ques_id[ques_id] == 0:
                    number_ques_per_paragraph_dict[file_id] += 1
                exist_ques_id[ques_id] += 1
                sent = content.get('sent')
                number_word = 0
                for sentence in sent:
                    number_word += len(sentence) 
                num_words_paragraphs.append(number_word)
               
            for i in number_ques_per_paragraph_dict.values():
                number_ques_per_paragraph.append(i)

# Example usage
file_path_train = '/home/minhhieu/Desktop/HMD_kaggle/data/feature_data/race_train_large_features.json'
file_path_dev = '/home/minhhieu/Desktop/HMD_kaggle/data/feature_data/race_train_large_features.json'
file_path_test = '/home/minhhieu/Desktop/HMD_kaggle/data/feature_data/race_train_large_features.json'

convert_json_file(file_path_train)
convert_json_file(file_path_dev)
convert_json_file(file_path_test)


# # Vẽ biểu đồ
# fig, axs = plt.subplots(2, 2, figsize=(20, 10))  # Tạo 4 biểu đồ con

# # Biểu đồ (a): Số từ trong câu trả lời
# axs[0, 0].hist(num_words_answers, bins=20, alpha=0.5, color='skyblue', edgecolor='black')
# axs[0, 0].set_xlabel('Number of Words in Answer')
# axs[0, 0].set_ylabel('Frequency')
# axs[0, 0].set_title('(a)')

# # Biểu đồ (b): Số từ trong câu hỏi
# axs[0, 1].hist(num_words_questions, bins=20, alpha=0.5, color='orange', edgecolor='black')
# axs[0, 1].set_xlabel('Number of Words in Question')
# axs[0, 1].set_ylabel('Frequency')
# axs[0, 1].set_title('(b)')

# # Biểu đồ (c): Số từ trong đoạn văn
# axs[1, 0].hist(num_words_paragraphs, bins=20, alpha=0.5, color='blue', edgecolor='black')
# axs[1, 0].set_xlabel('Number of Words in Paragraph')
# axs[1, 0].set_ylabel('Frequency')
# axs[1, 0].set_title('(c)')

# # Biểu đồ (d): Số lượng câu hỏi
# axs[1, 1].hist(num_words_distractor, bins=20, alpha=0.5, color='green', edgecolor='black')
# axs[1, 1].set_xlabel('Number word of Distractor')
# axs[1, 1].set_ylabel('Frequency')
# axs[1, 1].set_title('(d)')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def plot_lengths(lengths,column_name ,color='blue', title=None, xlabel=None, save_filename=None):
    # lengths = df[column_name].apply(lambda x: len(x.split()))

    plt.figure(figsize=(10, 6))
    sns.histplot(lengths, bins=20, kde=True, color=color)
    plt.xlabel(xlabel if xlabel else f'Number of Words in {column_name.capitalize()}')
    plt.ylabel('Frequency')
    if save_filename:
        plt.savefig(save_filename)
    plt.close()

def plot_answer_lengths(num_words_answers):
    plot_lengths(num_words_answers, 'correct', color='purple', xlabel='Number of Words in Answer', save_filename='answer_lengths_distribution.pdf')

def plot_question_lengths(num_words_questions):
    plot_lengths(num_words_questions, 'question', color='orange', xlabel='Number of Words in Question', save_filename='question_lengths_distribution.pdf')

def plot_paragraph_lengths(num_words_paragraphs):
    plot_lengths(num_words_paragraphs, 'context', color='blue', xlabel='Number of Words in Paragraph', save_filename='paragraph_lengths_distribution.pdf')

def plot_questions_per_paragraph(number_ques_per_paragraph):
    # questions_per_paragraph = df.groupby('context')['question'].count()

    plt.figure(figsize=(10, 6))
    sns.histplot(number_ques_per_paragraph, bins=20, kde=True, color='green')
    plt.xlabel('Number of Questions')
    plt.ylabel('Frequency')
    plt.title('Distribution of Questions per Paragraph')
    plt.show()
    # plt.savefig('questions_per_paragraph_distribution.pdf')
    # plt.close()

# Plot average lengths of incorrect answers
def plot_avg_incorrect_lengths(num_words_distractor):
    # avg_incorrect_lengths = df[['incorrect1', 'incorrect2', 'incorrect3']].apply(lambda x: pd.Series([len(option.split()) for option in x]), axis=1).mean(axis=1)
    
    plt.figure(figsize=(10, 6))
    sns.histplot(num_words_distractor, bins=20, kde=True, color='red')
    plt.xlabel('Average Number of Words in Incorrect Answers')
    plt.ylabel('Frequency')
    plt.savefig('avg_incorrect_lengths_distribution.pdf')
    plt.close()

# Assuming raw_train_df is your DataFrame
plot_answer_lengths(num_words_answers)
plot_question_lengths(num_words_questions)
plot_paragraph_lengths(num_words_paragraphs)
plot_questions_per_paragraph(number_ques_per_paragraph)
plot_avg_incorrect_lengths(num_words_distractor)
```
